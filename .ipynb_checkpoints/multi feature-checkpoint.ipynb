```python
import re
import csv
import glob
import numpy as np

import librosa
```

```python
path = '/home/ml-engineer/Documents/Dataset/data_speech_command'
```

```python
files_path = glob.glob(path + '/data_speech_commands_v0.01/*/*')
```

```python
len(files_path)
```

```python
files_path = files_path[52585:]
```

```python
header = 'filename chroma_stft rmse spectral_centroid spectral_bandwidth rolloff zero_crossing_rate'
for i in range(1, 21):
    header += f' mfcc{i}'
header += ' label'
header = header.split()
```

```python
file = open('data.csv', 'w', newline='')
with file:
    writer = csv.writer(file)
    writer.writerow(header)
```

```python
for file in files_path:
    path_list = file.split('/')
    
    y, sr = librosa.load(file)
    chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
    rmse = librosa.feature.rmse(y=y)
    spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
    spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    zcr = librosa.feature.zero_crossing_rate(y)
    mfcc = librosa.feature.mfcc(y=y, sr=sr)
    to_append = f'{path_list[-1]} {np.mean(chroma_stft)} {np.mean(rmse)} {np.mean(spec_cent)} {np.mean(spec_bw)} {np.mean(rolloff)} {np.mean(zcr)}'    
    for e in mfcc:
        to_append += f' {np.mean(e)}'
    to_append += f' {path_list[-2]}'
    file = open('data.csv', 'a', newline='')
    with file:
        writer = csv.writer(file)
        writer.writerow(to_append.split())
```

## Modeling

```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
ddata = pd.read_csv('data.csv')
```

```python
ddata.shape
```

```python
ddata.drop(['filename'], axis = 1, inplace=True)
```

```python
genre_list = ddata.iloc[:, -1]
```

```python
encoder = LabelEncoder()
```

```python
y = encoder.fit_transform(genre_list)
```

## Scaling the feature columns

```python
ddata.head()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(np.array(ddata.iloc[:, :-1], dtype = float))
```

## Train and Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20 ) 
```

```python
X_train.shape
```

```python
X_test.shape
```

## Classifier Keras

```python
from keras import models
from keras import layers
```

```python
model = models.Sequential()
```

```python
model.add(layers.Dense(256, activation = 'relu', input_shape = (X_train.shape[1],)))
```

```python
model.add(layers.Dense(128, activation = 'relu'))
```

```python
model.add(layers.Dense(64, activation = 'relu'))
```

```python
model.add(layers.Dense(30, activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam',
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs = 300, batch_size = 128)
```

```python
history_dict = history.history
```

```python
acc = history_dict['acc']
```

```python
sns.lineplot(x = list(range(300)), y = acc)
```
